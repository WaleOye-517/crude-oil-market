```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [27.0 , 5.0, 6.4, 33.1, 24.2, 4.3]
source= ['Coal_C(exj)','Renw(exj)','Hydro(exj)','Oil_C(exj)','Gas_C(exj)','Nuclear_C(exj)']
```

```python
colors = ['brown','green','blue', 'yellow', 'gray', 'red']
wp = {'linewidth': 1, 'edgecolor': 'black'}
explode = [0.1, 0, 0, 0.1, 0.1, 0.1]
def func(pct, allvalues):
    return '{:.1f}%'.format(pct/100*np.sum(allvalues))
plt.figure(figsize = (11,9))
plt.pie(x, labels = source, autopct = lambda pct : func(pct, x),textprops = dict(fontsize = 14,weight = 'bold'),explode = explode,
        wedgeprops = wp,colors = colors,startangle = 90)
plt.legend(title = 'Total Primary Consumption:',title_fontsize= 9,fontsize = 12,loc = 'center left', bbox_to_anchor = (1,0,0.5,1))
plt.title('World Primary Energy Consumption',loc = 'center',fontweight= 'bold', fontsize = 15)
plt.savefig('C:/Users/hp/Desktop/Data Science/Specialization Project/ass8.1.png', dpi = 300)
```

```python
fig, ax = plt.subplots(figsize=(12, 10), subplot_kw=dict(aspect="equal"))
investments = ['Oil&Gas','Renewables','Coal']

data =  [33.3, 55.6, 11.1]
wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=80,explode = explode)
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(investments[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

plt.title('World Global Investment In Power', fontsize = 25,loc = 'center',fontweight= 'bold')

plt.show()
```
