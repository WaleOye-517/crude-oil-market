```python
#plotting line plots with various line styles.
#use the help() function to find out the vailable line styles avaialble 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

help(plt.plot)
#look for linestyle or ls
```

```python
#create your data
x = [1,2,3,4,5]
y = np.arange(10,60,10)
```

```python
ls = ['-', '--', '-.', ':', '']
for l in ls:
    plt.plot(x,y, 'ko', ls = l)
    plt.title(str('Line style - {}'.format(l)), fontsize = 20, fontweight = 'bold')
    plt.show()
```

```python
#moving on to xticks
help(plt.xticks)
#look at docstrings below, what are you able to deduce?
```

```python
#Lets access our oil price data again and play around with the xticks
price = [43.12, 43.11, 42.82, 42.62, 43.4]
day = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri']

#day represents the categorical varaible that is located on the x-axis
#price is a continious variable that is located on the y-axis

#in other for us to be able to manipulate extra features in our charts, we can choose to create a complementary numerical variable for day
#it should be noted that this is not compulsory for line plots, but it can give you an edge.
x1 = np.arange(0, len(day))#what do you notice about this formula?
#we use the len finction to get the length of the day variable, we are creating a numerical range from 0 to length of the day varibale
```

```python
#since the x1 represents the day variable; it can be used as a replacement when plotting the charts and the same figure will be generated
#Let us compare and contrast

fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(day, price)
ax1.set_title('Using day')
ax2.plot(x1, price)
ax2.set_title('Using x1')

#spot any difference?
#just the labels right?
```

```python
#Let us use x1 to plot the chart and  add our own specifuc labels
plt.plot(x1, price, 'ro', ls = '-')
plt.xticks(x1, labels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'])
```

```python
#playing around with rotation of your xticks
rotation = ['horizontal', 'vertical', 45, 25, 135]
for rot in rotation:
    plt.plot(x1, price, 'ro', ls = '-')
    plt.xticks(x1, labels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], rotation = rot)
    plt.show()
    
#what do you notice from the charts below
#you can also increase font charts and change colours of the xticks by specifying
```

```python
#playing around with fontsize
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
ax1.plot(x1, price, 'bo', ls = '-')
ax1.set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], fontsize = 10, color = 'blue')
ax2.plot(x1, price, 'go', ls = '-')
ax2.set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], fontsize = 15, fontweight = 'bold', color = 'green')
ax3.plot(x1, price, 'ro', ls = '-')
ax3.set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], fontsize = 20, fontweight = 'bold', color = 'red', rotation = 'vertical')
```

# plt.fill_between()

The fill between is similar to the line plot and stackplot, infact, we can say it is the combination of the two. Let us see how it works.

```python
#using the oil price data
#let us first use the help(fill_between) to see how it works and the type of parameters and arguments it can take in
help(plt.fill_between)
```

```python
plt.fill_between(x1, price)
plt.xticks(x1, labels = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri'])
#what did you notice from the chart below?
#it brought out a square right, which is far from what we want
```

```python
#with the fill_between, we have to introduce the xlim and ylim
#xlim - preset the values of your x-axis(that us why it is important to use a numerical variable)
#ylim - preset the values of your y-axis(just like choosing an upper and lower boundary limit
help(plt.xlim), help(plt.ylim)
```

```python
#Let us re-plot our chart again using the xlim and ylim features
fig, ax = plt.subplots(figsize = (10,7))
plt.xlim(0,4)
plt.ylim(min(price)-0.2, max(price)+0.2)
ax.fill_between(x1, price)
plt.xticks(x1, labels = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri'])
#notice the difference?

```

```python
#let us now plot a line plot side by side with the fill_between plot
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,7))
#ax1.set_xlim(0,6)
ax1.set_ylim(min(price)-0.2, max(price)+0.2)
ax1.plot(x1, price)
ax1.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thur', 'Fri'])
#ax2.set_xlim(0,6)
ax2.set_ylim(min(price)-0.2, max(price)+0.2)
ax2.fill_between(x1, price)
ax2.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thur', 'Fri'])

#notice any peculiar diffrence?
```

```python
#creating a standard fill_between plot
lb = [2015, 2016, 2017, 2018, 2019, 2020]
v1 = [43, 70, 24, 28, 42, 23]
v2 = [39, 72, 52, 12, 21, 18]

fig, ax = plt.subplots(figsize = (18,12))
plt.ylim(0,80)
ax.plot(lb, v1, 'r', lw = 3)
ax.plot(lb, v1, 'wo', markersize = 10, markeredgecolor = 'red', markeredgewidth = 3, label = 'Oil and Gas Producer')
ax.plot(lb, v2, 'm', lw = 3)
ax.plot(lb, v2, 'wo', markersize = 10, markeredgecolor = 'magenta', markeredgewidth = 3, label = 'Oilfield Services')
ax.fill_between(lb, v1, color = 'red', alpha = 0.2, hatch = '.',in_layout = True)
ax.fill_between(lb, v2, color = 'magenta', alpha = 0.2, hatch = '.',in_layout = True)
for i,v in list(zip(lb, v1)):
    ax.text(i, v+3, str(v), fontsize = 15, fontweight = 'bold')
for i,v in list(zip(lb, v2)):
    ax.text(i, v-4, str(v), fontsize = 15, fontweight = 'bold')
plt.xticks(fontsize = 12, fontweight = 'bold')
plt.yticks(fontsize = 15)
ax.spines['right'].set_visible(False)                        
ax.spines['left'].set_visible(False)
#plt.tick_params(labeltop = False, labelright = True)                             
plt.grid(which = 'major', axis = 'y')
plt.ylabel('', fontsize = 12, fontweight = 'bold')
plt.legend(loc = 'upper right', fontsize = 14, markerfirst = True, fancybox = True)
plt.title('US OIL AND GAS BANKRUPTICES', fontsize = 20, fontweight = 'bold', loc = 'left')
plt.title('Source: Haynes and Boone, LLP.,Insolvency insider', fontsize = 10, loc = 'right')
plt.show()

```

```python
#look at the chart below, compare it to the one above
#what are the differences you observe
fig, ax = plt.subplots(figsize = (18,12))
plt.ylim(0,80)
ax.plot(lb, v1, 'r', lw = 3, label = 'Oil and Gas Producer')
ax.plot(lb, v2, 'm', lw = 3, label = 'Oilfield Services')
plt.yticks(fontsize = 15)
plt.ylabel('', fontsize = 12, fontweight = 'bold')
plt.legend(loc = 'upper right')
plt.title('US OIL AND GAS BANKRUPTICES', fontsize = 20, fontweight = 'bold', loc = 'left')
plt.title('Source: Haynes and Boone, LLP.,Insolvency insider', fontsize = 10, loc = 'right')
plt.show()

```
