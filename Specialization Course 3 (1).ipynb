# Working with Bar Charts


Bar charts, just like line charts, are similar to line plots in the sense that they take a categorical variable on the x-axis and a continious variable on the y-axis.

For this lecture series, you will be learning about four different kinds of bar chart and their styles of formatiing:
1) Basic Bar chart
2) Horizontal Bar chart
3) Stacked Bar Chart
4) Multiple Bar Chart

# Basic Bar Chart (plt.bar)

```python
#using the help function to see the kind of variables it can take in
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

help(plt.bar)
```

```python
#creating the data to be used
#Let us work with  a dataframe this time around
country = ['Nigeria', 'USA', 'Saudi', 'Russia', 'China', 'Canada']
O_P = [10000, 60000, 50000, 40000, 20000, 30000]#O_P stands for oil production(this are nit the correct values)
df = pd.DataFrame({'Country': country, 'Production': O_P})

df

```

```python
plt.bar(df['Country'], df['Production'])
```

```python
#you can try to increase/decrease the width of the bars
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,7))
ax1.bar(df['Country'], df['Production'], width = 0.3)
ax2.bar(df['Country'], df['Production'], width = 1)
```

```python
plt.bar(df['Country'], df['Production'], align = 'edge')
```

```python
plt.bar(df['Country'], df['Production'], align = 'center')
```

```python
#just like the line plots that you have treated, you can also play around with the xticks and the labels
#you have to create a numerical variable for the x-axis that will give you more control over the charts
x = np.arange(0, len(df))
x
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,7))
ax1.bar(x, df['Production'])
ax2.bar(df['Country'], df['Production'])
#you will still get the same chart, just different labels, let us correct this below for the x variable as we add labels to the plot
```

```python
fig, ax = plt.subplots()
ax.bar(x, df['Production'])
plt.xticks(x, labels = df['Country'])
```

```python
#you can also close up the space, control maximum and minimum values and allow the bar plots start exactly from 0 using the xlim and ylim
fig, ax = plt.subplots()
plt.xlim(min(x), max(x))#since it a numpy array
ax.bar(x, df['Production'], color = 'navy')
plt.xticks(x, labels = df['Production'])
#did you like the way the plot came out below? NO right, so you definitely want to correct that. Check the code below
```

```python
#correct the xlim by playing around
fig, ax = plt.subplots()
plt.xlim(min(x)-0.5, max(x)+0.5)#I added and subtarcted 0.5 from the maximun and minimum points respectively
ax.bar(x, df['Production'], color = 'navy')
plt.xticks(x, labels = df['Country'])

#is the chart below better? definitely
#you can also adjust the values, depending on how you want your bar plots to look like
```

```python
#you can also adjust the position of the xticks, this is convenient when you are using a numerical variable
#just like your x varaibele in this case unlike using the original labels

fig, ax = plt.subplots()
ax.bar(x, df['Production'])
plt.xticks(x+0.5, labels = df['Country'])

#the adjustment of the position doesn't have a lot of use for a basic bar plot, but it plays an important role in multiple bar charts
#take a look at the code below, if you plan on adjusting the xticks, you can also adjust the your plots on the x-axis
```

```python
fig, ax = plt.subplots()
ax.bar(x+3, df['Production'])
plt.xticks(x+3, labels = df['Country'])
```

```python
#let us play around with colors
#just like your line plots, you can always change the color of the plots
fig, ax = plt.subplots(5, 1, figsize=(5,20))
colors = ['green', 'red', 'navy', 'yellow', 'gold']
for axs, color in zip(ax, colors):
    axs.bar(df['Country'], df['Production'], color = color)
    #plt.tight_layout(pad=0.4, w_pad=1.0, h_pad=1.0)
    
```

```python
#you can also try to change the colors of each plot depending on the colors
df = pd.read_excel('/Users/Hp/Desktop/Breakeven_Oil_Prices.xlsx', sheet_name = 'fiscal')#change the directory

colors = []#creating a list of colors based on conditions
for price in df['breakeven']:
    if price <= 50:
        tt = 'tab:green'
    elif price <= 100:
        tt = 'tab:blue'
    else:
        tt = 'tab:red'
    colors.append(tt)
    
df['colors'] = pd.DataFrame(colors)#append the new color list as a dataframe in the original dataframe

df = df.sort_values(by = 'breakeven', ascending = True)# sort the values in increasing order
colors = [i for i in df['colors']] #create a list of the sorted dataframe colors

x1 = np.arange(0, len(df))
fig, ax = plt.subplots(figsize = (10,7))
ax.bar(x1, df['breakeven'], color  = colors)
ax.spines['right'].set_visible(False)             
# ax.spines['top'].set_visible(False)             
ax.spines['left'].set_visible(False)             
ax.spines['bottom'].set_visible(False)
plt.xticks(x1, labels = df['country'], fontsize = 15, rotation = 45)
plt.yticks(fontsize = 10)
plt.title('Fiscal Breakeven Prices', fontsize = 20, fontweight = 'bold', loc = 'left')
plt.title('Source: knoema.com', loc = 'right')
for i, v in list(zip(x1, df['breakeven'])):
    ax.text(i-0.3, v+5, str('${:.1f}'.format(v)), fontsize = 13)
```
