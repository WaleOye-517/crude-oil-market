# Pie Charts(plt.pie)

In this module, we will cover the basics of pie charts and dive into how they can be are used in the real world. We will explore basic features that will help you bring more life into the normal pie chart.

```python
#The first thing just like we always do is to explore the documentation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
help(plt.pie)
```

```python
#creating instances
x = [20, 30, 40, 50]
label = ['NYU', 'DC', 'SHC', 'LVG']
```

```python
plt.pie(x, labels = label)
plt.title('A simple Pie Chart')
plt.show()
```

```python
#Adding the percentage(pct) function
#increasing size of chart
#increasing textfonts using texprops parameter
def func(pct, allvalues):
    return '{:.1f}%'.format(pct)

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20))
plt.title('A Simple Pie Chart', fontsize = 25)
```

Let us a take a detour and try to understand the def func(pct, allvalues) function.

The function above takes in two arguments, the pct value which is automaitcally set as a percentage function and the values of the chart(in this case, the 'x' list).
After the creation of the function, it is then called with the autopct parameter in plt.pie and it is passed in using a lambda function.(You can read up on how a lambda fucntion works, but you all actually don't need to, once you understand how the autopct parameter takes in values, you are good to go.)

```python
#What if we just want our values and not percentages
#then we can edit the value we return as seen below
#it is the return value that is edited depending on what you want


def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))#key

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20))
plt.title('A Simple Pie Chart', fontsize = 25)
```

```python
#Let us make things more fun 
#now we will add the explode parameter

#tips for the explode parameter
#1. Must contain the same no of values as the x and labels parameter
#2. Can be preset an passed in as a list

explode = [0.1, 0.1, 0.1, 0.1]
def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))

plt.figure(figsize = (10,7))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20), explode = explode)
plt.title('Do you want a piece of Pie?', fontsize = 25)
```

```python
#you can also prest the colors you want and add wedgeprops

colors = ['tab:red', 'tab:blue', 'tab:brown', 'tab:purple']

wp = {'linewidth': 3, 'edgecolor': 'black'}

explode = [0.1, 0.1, 0.1, 0.1]

def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20),
       colors = colors,
       explode = explode,
       wedgeprops = wp)
plt.legend(loc = 'center left', bbox_to_anchor = (1,0,0.5,1), fontsize = 20)
plt.title('Do you want a Fancy piece of pie?', fontsize = 25)
```

```python
#You can also specify and highlight by playing around with the explode parameter
colors = ['tab:red', 'tab:blue', 'tab:brown', 'tab:purple']

wp = {'linewidth': 3, 'edgecolor': 'black'}

explode = [0, 0, 0, 0.2]

def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20),
       colors = colors,
       explode = explode,
       wedgeprops = wp)
plt.title('Do you want a Fancy piece of pie?', fontsize = 25)
```

Now you have seen how to tranform a chart from a simple pie chart to a chart worth publishing, there are a lot of other customizations that you can do, feel free to use the internet and the help documentation to create beautiful pie charts that depicts the information you are trying to pass across.

# Donot Plot

Donot plot, just like the names imply is a pie chart with a circle in the center. To create a donot plot, you first have to create a pie chart and then add extra customizations

```python
#We will start by creating a basic pie chart

colors = ['tab:red', 'tab:blue', 'tab:brown', 'tab:purple']

wp = {'linewidth': 3, 'edgecolor': 'black'}


def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20),
       colors = colors,
       wedgeprops = wp)
plt.title('Do you want a Fancy piece of pie?', fontsize = 25)
```

```python
#For us to create a donut plot, we have to use the plt.Circle method
#help(plt.Circle)
help(plt.Circle)
```

```python
#Now, we can introduce plt.Circle and pass it as a variable

my_circle = plt.Circle( (0,0), 0.8, color = 'white', ec ='black', lw = 2)
colors = ['tab:red', 'tab:blue', 'tab:brown', 'tab:purple']

wp = {'linewidth': 3, 'edgecolor': 'black'}

def func(pct, allvalues):
    return '${:.0f}'.format(pct/100*np.sum(allvalues))

plt.figure(figsize = (12,10))
plt.pie(x, labels = label, autopct = lambda pct : func(pct, x),
       textprops = dict(fontsize = 20),
       colors = colors,
       wedgeprops = wp)
plt.gcf().gca().add_artist(my_circle)#This is key, take note of this line, without it there will be no effect
plt.legend(loc = 'center left', bbox_to_anchor = (1,0,0.5,1), fontsize = 20)
plt.title('Do you want a Fancy piece of pie?', fontsize = 25)
```

Now, you have seen how to create a basic donut plot, take note of the radius parameter in plt.Circle and play around with it to see the effect it makes on the chart.

You should know that all methods and effects that can be made with a pie chart can also be made with a donut plot.
