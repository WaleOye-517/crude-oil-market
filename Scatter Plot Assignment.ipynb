```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ass = pd.read_excel('C:/Users/hp/Desktop/Data Science/Specialization Project/Scatterplot Practice.xlsx',sheet_name = 'Sheet2')
```

```python
plt.rc('font', family = 'serif')
fig, ax = plt.subplots(figsize =(10,7))
ax.scatter(ass['Oil_C(mbbl)'], ass['C02(mt)'],color = 'white', edgecolor = 'tab:red', s = 50, lw = 3)
ax.plot([ass['C02(mt)'].min(), ass['C02(mt)'].max()], [ass['C02(mt)'].min(), ass['C02(mt)'].max()],
        '--', color = 'black', lw = 4)
plt.ylabel('Emissions')
plt.xlabel('Oil Consumption')
plt.grid('major', axis = 'y')
plt.tick_params(labelright = True)
plt.title('Oil Consumption vs Emissions', fontsize = 15)
```

```python
ass[['C02(mt)', 'Oil_C(mbbl)']].corr()['C02(mt)']
```
