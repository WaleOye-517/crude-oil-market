# Scatter Plots(plt.scatter)


Scatter plots are very similar to line plots, the only difference is that they the line traces are not present.
the tasks that can be accomplished with scatter plots can also be done with line plots, but there are cases when you will choose to use one over the other.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(np.random.randint(0,253.5,size=(250, 2)).astype('float'), columns = ['Sales', 'Profits'])
```

```python
help(plt.scatter)
'''just like in previous sessions, let us use the help function 
    to see the kind of parameters plt.scatter can take '''
```

```python
#creating a simple scatter plot
'''One thing to note is that scatter plots should always be used with numerical variables
both on the x and y axis, and unlike line plots that can accept categorical numerical variables
on the x-axis, it is not advisable to use a it for a scatter plot since what we seek when using a 
scatter plot is correlation between two variables'''
plt.scatter(df['Sales'], df['Profits'])



print('The chart below is poorly correlated as the points are far apart from one another')
```

```python
'''Let us create a sorted list and see how the plot above changes
    there are various ways in which one might decide to use scatter plot
    but they are mainly used in showing correlation between two variables
    hence this makes scatter plots useful in the process of data science for
    machine learning as they can help in feature selection'''
new_list1 = []
new_list2 = []
for i in df['Sales']:
    new_list1.append(i)

for i in df['Profits']:
    new_list2.append(i)
```

```python
'''The sorted list has  been creates and this is what we will be using below to plot our chart'''
new_list1.sort(), new_list2.sort()
```

```python
plt.scatter(new_list1, new_list2)
print('This is an example of a correlated variable, as you can see the points are closely packed to each other')
```

```python
'''Do not be misguided by the example above, the data used for this example was created using random samples
and that is why I sorted them to generate the chart above, in an real situation, sorting your data in other to
get a correlated plot is very wrong!

Let us now play around with some formatting'''

fig,(ax1,ax2,ax3, ax4) = plt.subplots(1,4, figsize = (10,7))
ax1.scatter(new_list1, new_list2, s= 20, color = 'magenta')
ax2.scatter(new_list1, new_list2, s= 40, color = 'red')
ax3.scatter(new_list1, new_list2, s= 60, color = 'blue')
ax4.scatter(new_list1, new_list2, s= 80, color = 'black')

```

```python
plt.rc('font', family = 'serif')
fig, ax = plt.subplots(figsize =(10,7))
ax.scatter(scc['Production'], scc['Consumption'],color = 'white', edgecolor = 'tab:red', s = 100, lw = 3)
ax.plot([scc['Production'].min(), scc['Production'].max()], [scc['Production'].min(), scc['Production'].max()],
        '--', color = 'black', lw = 4)
plt.ylabel('Consumption')
plt.xlabel('Production')
plt.grid('major', axis = 'y')
plt.tick_params(labelright = True)
plt.title('Production vs Consumption', fontsize = 15)
```

From the chart above, I created a diagonal line that acts as a regressor line showing how well the variables are correlated with each other, if the points are closely packed together, it can be said that they are well correlated. There is a way to show you the correlation value between the two variables:

```python
scc[['Production', 'Consumption']].corr()['Consumption']

'''From the results below it can be seen that Production as a 0.996005 value with Consumption
    and Consumption has 1.0 correlation with itself, which of course is to be expected.
    
    The point of you knowing the correlation value is important when it comes to feature selection in data
    analytics, you can easily choose variables that are well correlated with the target feature simply by
    plotting your scatter plot.
    
    NOTE: Correlation does not always imply causation i.e The fact that two variables are well correlated does
    not imply you can use one of the variables to predict the other variable.'''
```

```python
plt.rc('font', family = 'serif')
fig, ax = plt.subplots(figsize =(10,7))
ax.scatter(df['Sales'], df['Profits'], color = 'white', edgecolor = 'tab:red', s = 100, lw = 3)
ax.plot([df['Sales'].min(), df['Sales'].max()], [df['Sales'].min(), df['Sales'].max()],
       'k--', lw = 3)
plt.ylabel('Profits')
plt.xlabel('Sales')
plt.grid('major', axis = 'y')

'''From the charts below, we can see poorly correlated variables
    let us use the df.corr() method to check out the value
    
    Note: the corr() method only works on numerical variable and that is why
    scatter plot is use for numerical variables only'''

df.corr()['Sales']
```

# Plotting 3 varibles at once with a Scatter plot 

Scatter plot is capable of taking in three varibles and showing the relationship they have on each other, this type is not common but used in certain situations, we will take a look at some example below, but first let us use the help(plt.scatter) to understand how this feature works

```python
help(plt.scatter)
'''the parameter c is what we need here, from the documentation below, we can see that
    it takes in an array and it is originally meant to take in a list of colors
    
    We can use this to our advantage and create a plot that shows the effect a third variable
    has on the other two variables with a change in color sequentially allong the diagonal'''
```

```python
plt.rc('font', family = 'serif')
fig, ax = plt.subplots(figsize =(10,7))
ax.scatter(scc['Production'], scc['Consumption'], label = 'Oil price',
           c = scc['Oil Price'], s = 100, lw = 3, cmap = 'viridis')
ax.plot([scc['Production'].min(), scc['Production'].max()], [scc['Production'].min(), scc['Production'].max()],
        '--', color = 'black', lw = 4)
plt.ylabel('Consumption')
plt.xlabel('Production')
plt.grid('major', axis = 'y')
plt.tick_params(labelright = True)
plt.title('Production vs Consumption', fontsize = 15)
#plt.text(35000, 95000, str('{}'.format(scc[['Production', 'Consumption']].corr()['Consumption'])))
```

```python
'''Here is a better way to plot it to show more details, we use pandas plotting technique
    this time around so as to make things easier.
    
    As you can see from the code below, we are able to take in extra parameters using this style such as colorbar
     which is useful in indicating the sequence of the third variable.'''

ax = scc.plot(kind="scatter", x="Production", y="Consumption",
             alpha=0.8,  s = 100,
             figsize=(10,7),    c="Oil Price", cmap=plt.get_cmap("rainbow"),
             colorbar=True,    sharex=False) 
ax.plot([scc['Production'].min(), scc['Production'].max()], [scc['Production'].min(), scc['Production'].max()],
        '--', color = 'black', lw = 2)
ax.spines['right'].set_visible(False)                        
ax.spines['left'].set_visible(False) 
plt.legend(fontsize = 15)
plt.grid(which = 'major', axis = 'y')
plt.ylabel('Oil Production in Thousand Barrels', fontsize = 15, fontweight = 'bold')
plt.xlabel('Oil Production in Thousand Barrels', fontsize = 15, fontweight = 'bold')
plt.title('Relationship Between Oil Production, Oil Consumption and Oil Price', fontsize = 15, fontweight = 'bold')
```

```python
'''From the diagram above, we can see how three variables were represented with 
    a scatter plot. To be more specific, 3 Numerical variables.
    
    The chart tells the story of how oil production has increased with oil consumption over
    the years and this has led to an increase in oil price, although oil price has started to
    see a decline along the way'''
```
