```python
def month(x):
    if x == '12':
        return 'Dec'
    elif x == '01':
        return 'Jan'
    elif x == '02':
        return 'Feb'
    elif x == '03':
        return 'March'
    elif x == '04':
        return 'April'
    elif x == '05':
        return 'May'
    elif x == '06':
        return 'June'
    elif x == '07':
        return 'July'
    elif x == '08':
        return 'Aug'
    elif x == '09':
        return 'Sep'
    elif x == '10':
        return 'Oct'
    elif x == '11':
        return 'Nov'
```

```python
crude_oil.dropna(inplace=True)
```

```python
crude_oil['date2'] = crude_oil['Date']
crude_oil = crude_oil.set_index(crude_oil['date2'])
crude_oil['monthno'] = crude_oil['Date'].str.split('-', 2).str[1]
crude_oil['month'] = crude_oil['monthno'].apply(month)
crude_oil['dayno'] = crude_oil['Date'].str.split('-', 2).str[2]
crude_oil['year'] = crude_oil['Date'].str.split('-', 2).str[0]
crude_oil['main'] = crude_oil['month']+' '+crude_oil['dayno']
```
