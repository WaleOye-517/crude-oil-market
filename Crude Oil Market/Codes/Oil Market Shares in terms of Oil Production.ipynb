```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
```

```python
Contributors = ['OPEC', 'Non-OPEC', 'OPEC+', 'US', 'Saudi Arabia', 'Russia','Canada']
Share = [37.4, 62.6, 56, 17.9, 12.4, 12.1,5.9]
df = pd.DataFrame({'Contributors': Contributors, 'Share': Share})
x=np.arange(0,len(df))
color = ['yellow','blue','brown','green','violet','turquoise','purple']
```

```python
width =0.5
fig, ax = plt.subplots(figsize = (12,10))
plt.bar(df['Contributors'],df['Share'], width=width,align = 'center',color = color)
plt.grid(which = 'major', axis = 'y', color = 'black')
plt.xticks(df['Contributors'], rotation = 45, fontsize = 15,fontweight = 'bold')
for i, v in list(zip(df['Share'], df['Contributors'])):
   ax.text(v, i + .35, str('{}%'.format(i)),fontsize = 16, fontweight='bold',rotation =10)
plt.text(3, 45,'OPEC shares already include Saudi Arabia shares\nOPEC+ shares include OPEC shares and Saudi Arabia\nNon-OPEC shares include US,Russia and Canada',fontsize = 13,fontweight = 'bold')
plt.title('Oil Market Shares in terms of Oil Production', fontsize = 20, fontweight = 'bold', loc = 'left')
plt.title('Source: BP Statistical Review 2020', fontsize = 10, loc = 'right')
plt.tight_layout()
plt.show()
```

```python
Year = [1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
Non_OPEC = [115, 185, 260, 320,390,480,500,530,600]
OPECplus = [300,385,500,615,800,900,1140,1280,1400]
df = pd.DataFrame({'Year': Year, 'Non-OPEC': Non_OPEC,'OPEC+':OPECplus})
x=np.arange(0,len(df))
```

```python
fig, ax = plt.subplots(figsize = (12,10))
ax.stackplot(df['Year'],df['Non-OPEC'], df['OPEC+'],colors = ['b','r'],labels= ('Non-OPEC','OPEC+'))
plt.yticks(fontsize = 15)
plt.xticks(df['Year'], fontsize = 15)
ax2 = ax.twinx()
ax2.stackplot(df['Year'],df['Non-OPEC'], df['OPEC+'],colors = ['b','r'],labels= ('Non-OPEC','OPEC+'))
plt.yticks(fontsize = 15)
ax.grid(which = 'major', axis = 'y')
plt.xticks(df['Year'], fontsize = 15)
plt.legend(loc = 'upper left', title = 'Total World Reserve by:',fontsize = 15,ncol = 2,title_fontsize = 20)
plt.title('Increase In Total World Reserves' ,fontweight='bold',fontsize = 25,loc = 'left')
plt.title('Source: BP Statistical Review 2020', fontsize = 10, loc = 'right')
plt.show()
```
