# Creating line plots - plt.plot()

```python
#use the help function to find out more
help(plt.plot)
```

```python
#creating the data
x = [1,2,3,4,5]
y = np.arange(10,60,10)
#x and y must have the same no of values
#for line plots, it can take both continious values as x and y, but it is advisable
#that the x-axis should be a categorical variable 

#Note: there are numeric categorical variable like date, age, year that can act as numeric and categorical varibles in both sense
```

```python
#basic plotting
plt.plot(x,y)
```

```python
#basic pooting with colors
plt.plot(x,y, 'r')
#other available colors include : 'b', 'g', 'k'
#use the help function to get more details
```

```python
#basic plotting with colors and markers
plt.plot(x, y, 'ro')
plt.show()
plt.plot(x,y, 'g+')
plt.show()
plt.plot(x,y, 'y*')
plt.show()
```

```python
#learn how to use subplots
#subplots basically allow you to compare or plots to charts beside one another
help(plt.subplots)
```

```python
#Comparing the plots
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x,y,'r')
ax2.plot(x,y,'ro')
#what difference do you notice between the two charts below, let us see if we can combine both in our next line of code
```

```python
plt.plot(x,y,'r')
plt.plot(x,y,'ro')
#What do you see from the results below?, this is the beginnig of manipulation of our charts to get what we want
```

```python
#let us try the charts with various features
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize = (10,7))
ax1.plot(x,y,'r--')#you can specify the lind of lines, iin this case, dashed lines was specified
ax2.plot(x,y,'g', lw = 6)#thicker lines compared to other line plots
ax3.plot(x,y,'wo',markersize = 10, markeredgecolor = 'magenta', markeredgewidth = 3)#adding various specialization to the markers
ax4.plot(x,y, color = 'gold', lw = 10, alpha = 0.25)#controls the transparency of the plot
```

Enough with the play, let us see a perfect illustration of how line plots are used in the real-life

A perfecet example willl be in the monitioring of prices.

The data below contains oil price data for August, the code below shows the customization of the the chart using line charts

```python
price = [43.12, 43.11, 42.82, 42.62, 43.4]
day = ['Mon', 'Tue', 'Wed', 'Thurs' 'Fri']

x1 = np.arange(0, len(day)+1)#creating a numerical variable that will replace the day variable, this will give us extra manipulation
x1
```

```python
plt.rc('font', family = 'serif')#specifying font type
fig, ax  = plt.subplots(figsize = (20,14))
plt.xlim(0,4)#controlling the limits of the x-axis; setting a minimun and maximun boundary
plt.ylim(42.2,43.6)#controlling the limits of the y-axis; setting a minimun and maximun boundary
plt.grid(which = 'major', axis = 'y', color = 'black')#creating the grid pressnt in the chart
ax.plot(x1, price, 'g', lw = 7)
ax.plot(x1, price, 'wo', markersize = 15, markeredgecolor = 'green', markeredgewidth = 5, label = 'Oil price')
ax.plot(x1[-1], price[-1], 'ro', markersize = 30)#creating the red colored spot in the chart below; specifying extra details in the chart
plt.xticks(x1, labels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], fontsize = 15, fontweight = 'bold', color = 'black')#setting the ticks ''(this is why it was important to create the x1 variable instead of using the day variable directly) and creating lables of my choice        
plt.yticks(fontsize = 15, fontweight = 'bold', color = 'black')
for i, v in list(zip(x1, price)):#(this allow me to show the price each day, try the code below)
    if v == np.nan:
        pass
    else:
        ax.text(i,v+0.03, str('${}'.format(v)), fontsize = 15, color = 'black')
plt.legend(fontsize = 20, loc = 'upper left')
plt.title('Test1', fontsize = 30, fontweight = 'bold', color = 'black')
#from the codes above, you will see that fontsize, fontweight, and color were specified in some lines of code
#this is important when dealing with visiblity of essential details, you can change it depending on your choice.
```

```python
plt.rc('font', family =  'elephant')
fig, ax  = plt.subplots(figsize = (30,21))
plt.xlim(0,4)#controlling the limits of the x-axis; setting a minimun and maximun boundary
plt.ylim(42.2,43.6)#controlling the limits of the y-axis; setting a minimun and maximun boundary
plt.grid(which = 'major', axis = 'y', color = 'black')#creating the grid pressnt in the chart
ax.plot(x1, price, 'k--', lw = 7)
ax.plot(x1, price, 'bo', markersize = 20, markeredgecolor = 'black', markeredgewidth = 5)
ax.plot(x1[-1], price[-1], 'yo', markersize = 30)#creating the red colored spot in the chart below; specifying extra details in the chart
plt.xticks(x1, labels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'], fontsize = 15, fontweight = 'bold', color = 'red')#setting the ticks ''(this is why it was important to create the x1 variable instead of using the day variable directly) and creating lables of my choice        
plt.yticks(fontsize = 15, fontweight = 'bold', color = 'red')
plt.title('TEST2', fontsize = 30, fontweight  ='bold', color = 'blue',loc = 'right')
plt.title('TEST2', fontsize = 30, fontweight  ='bold', color = 'brown', loc = 'left')
plt.title('TEST2', fontsize = 30, fontweight  ='bold', color = 'red')

#can you point out the differences with the first plot
#identify the chnages in the code
```
